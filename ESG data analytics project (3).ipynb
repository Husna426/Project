```python
import pandas as pd
import pandas as pd
import seaborn as sns
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
#Load data
df = pd.read_csv('esg data.csv')
print(df)
```

```python
#Looking at the data
df.head()
```

```python
# Check if there are any missing values in the DataFrame
missing_data = df.isnull().sum()

print("Missing Data:")
print(missing_data)

if missing_data.any():
    print("\nThere are missing values in the dataset.")
else:
    print("\nNo missing values found in the dataset.")
```

```python
df = df.dropna()
```

```python
print(df.columns)
```

```python
#Dropping duplicates
df.drop_duplicates()
```

```python
#Data exploration
print(df.describe())
```

```python
#Checking the data type of the columns
print(df.dtypes)
```

```python

# Plot distribution of ESG scores
plt.figure(figsize=(16, 10))
plt.hist(df['environment_score'], bins=20, color='blue', alpha=0.7, label='Environment Score')
plt.hist(df['social_score'], bins=20, color='green', alpha=0.7, label='Social Score')
plt.hist(df['governance_score'], bins=20, color='red', alpha=0.7, label='Governance Score')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Distribution of ESG Scores')
plt.legend()
plt.show()
```

```python
# Convert dates to a consistent format ('%d-%m-%Y')
df['last_processing_date'] = pd.to_datetime(df['last_processing_date'], errors='coerce', dayfirst=True)

# Check if there are any NaT (Not a Time) values, indicating parsing errors
if df['last_processing_date'].isnull().values.any():
    print("Some dates could not be parsed properly. Check for inconsistencies in the date format.")
    
    # Print the rows with unparsable dates for manual inspection
    print(df[df['last_processing_date'].isnull()])

    # If you identify inconsistencies, correct them and reattempt parsing
    # For example, if '7/6/2022' should be '07-06-2022', you can replace '/' with '-'
    df.loc[df['last_processing_date'].isnull(), 'last_processing_date'] = df.loc[df['last_processing_date'].isnull(), 'last_processing_date'].str.replace('/', '-')
    df['last_processing_date'] = pd.to_datetime(df['last_processing_date'], errors='coerce', dayfirst=True)

# Drop rows with missing or incorrect dates
df.dropna(subset=['last_processing_date'], inplace=True)

# Convert all dates to the desired format ('%d-%m-%Y')
df['last_processing_date'] = df['last_processing_date'].dt.strftime('%d-%m-%Y')

# Print the DataFrame to verify the changes
print(df)
```

```python
df['last_processing_date'] = pd.to_datetime(df['last_processing_date'], format='%d/%m/%Y')

# Print the DataFrame to verify the changes
print(df)
```

```python
# Compare ESG scores across industries
plt.figure(figsize=(16, 10))

# Rest of the code remains the same
df.groupby('industry')[['environment_score', 'social_score', 'governance_score']].mean().plot(kind='bar')
plt.xlabel('Industry', fontsize=14)
plt.ylabel('Mean Score', fontsize=14)
plt.title('Mean ESG Scores by Industry', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


```

```python
#Trends of ESG score overtime
if 'last_processing_date' in df.columns:
    df['last_processing_date'] = pd.to_datetime(df['last_processing_date'], format='%m/%d/%Y')
    plt.figure(figsize=(16, 10))
    plt.plot(df['last_processing_date'], df['environment_score'], marker='o', label='Environment Score')
    plt.plot(df['last_processing_date'], df['social_score'], marker='o', label='Social Score')
    plt.plot(df['last_processing_date'], df['governance_score'], marker='o', label='Governance Score')
    plt.xlabel('Date')
    plt.ylabel('Score')
    plt.title('Trend of ESG Scores Over Time')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("No date column for trend analysis.")
```
