```python
#Load data
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('Customer-Churn-Records.csv')
print(df)
```

```python
#Looking at the data
df.head()
```

```python

# Check if there are any missing values in the DataFrame
missing_data = df.isnull().sum()

print("Missing Data:")
print(missing_data)

if missing_data.any():
    print("\nThere are missing values in the dataset.")
else:
    print("\nNo missing values found in the dataset.")
```

```python
#Removing null values from the data
df = df.dropna()
```

```python
print(df.columns)
```

```python
#Dropping duplicates
df.drop_duplicates()
```

```python
#Data exploration
print(df.describe())
```

```python
#Checking the data type of the columns
print(df.dtypes)
```

```python
df.hist(figsize=(10, 8))
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Visualize correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Age', bins=20, kde=True)
plt.title('Distribution of Customer Age')
plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='HasCrCard')
plt.title('Credit Card Ownership')
plt.xlabel('Has Credit Card (1=yes, 0=no)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='EstimatedSalary', hue='Exited')
plt.title('Age vs. Estimated Salary (Churn)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend(title='Churn')
plt.show()
```
